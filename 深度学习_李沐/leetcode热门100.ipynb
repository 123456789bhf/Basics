# 11盛水最多的容器

```python
#超出时间限制
class Solution:
    def maxArea(self, height: List[int]) -> int:
        maxvalue=0
        for first in range(len(height)):
            for last in range(first+1,len(height)):
                if maxvalue<min(height[first],height[last])*(last-first):
                    maxvalue=min(height[first],height[last])*(last-first)

        return maxvalue
```

```python
#超出内存限制
def threeSum(nums):
        #如果列表是空集或者列表的元素的个数小于3，那么返回[]、
        #对列表排序，并且去掉重复元素
        #遍历整个列表
        #如果num[i]>0:因为樱井牌序号，所以后面不可能与三个数加和等于0，直接返回结果
        #另左指针L=i+1,右指针R=n-1，当L<R的时候执行循环：
        #当nums[i]+ums[R]+nums[L]==o，将[L,i，R]添加到结果列表中，继续循环
        #如果和大于0，说明nums[R]太大，R左移动
        #如果和小于0，说明nnums[L]太小，L右移动
        alist=list(set(nums))
        stop=False
        if alist==[] or len(alist)<3:
            return []
        else:
            i=0
            result=[]
            while i<len(alist) and not stop:
                #如果大于0，那么跳出循环，因为不可能找到
                if alist[i]>0:
                    stop=True
                else:
                    L=i+1
                    R=len(alist)-1
                    while L<R:
                        if alist[i]+alist[R]+alist[L]==0:
                            result.append([nums[L],nums[i],nums[R]])
                            L=L+1
                            R=R-1
                        elif alist[i]+alist[R]+alist[L]>0:
                            R=-1
                        else:
                            L=L+1
                    
            return result
print(threeSum([-1,0,1,2,-1,-4]))
```

## 3.无重复字符串的最长字串

```python
def lengthOfLongestSubstring(s):
    """
    :type s: str
    :rtype: int
    """
    has1={}
    for string in s:
        if string not in has1.keys():
            has1[string]=1
        else:
            has1[string]=has1[string]+1
    return max(list(has1.values()))
print(lengthOfLongestSubstring('aasdsaa'))
```

## 438 找到字符串中所有的异序词

```python
class Solution(object):
    def findAnagrams(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: List[int]
        """
        if len(s)<len(p):
            return []
        length=len(p)
        res=[]  #将所有三个值都放入一个列表中
        result=[] #储存结果
        alist=s.split()
        for i in range(len(alist)-length+1):
            res.append(alist[i:i+length])
        for l in range(len(res)):
            goal=p.split()
            for i in range(len(res[l])):
                if res[l][i] in goal:
                    goal.remove(res[l][i]) #移除i元素
            if goal==[]:
                result.append(l)
        return result



                 
             

            
```
